```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/Users/ravi/Desktop/flight/data/raw/jan2025.csv')
df.head()
print(df.shape)
```

```python
print(f"rows' = {df.shape[0]}, 'columns' = {df.shape[1]}")
print(df.columns)
total_flights = len(df)
cancelled_flights = df.CANCELLED.sum()
percent_cancllled = (cancelled_flights / total_flights) * 100
print("Percentage of cancelled flights: ", percent_cancllled)
```

```python


df.MKT_UNIQUE_CARRIER.value_counts()

df.MKT_UNIQUE_CARRIER.value_counts().plot.pie(autopct='%1.1f%%')
plt.show()

```

```python
cancellations_per_airline = df.groupby('MKT_UNIQUE_CARRIER')['CANCELLED'].mean() * 100
carrier_cancel_dict = cancellations_per_airline.to_dict()

print(carrier_cancel_dict.keys())
print(carrier_cancel_dict.values())
for key in carrier_cancel_dict.keys():
    print(f"{key} has {carrier_cancel_dict[key]:.2f}% cancellation rate")


```
